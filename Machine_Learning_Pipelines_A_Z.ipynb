```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.feature_selection import chi2,SelectKBest
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

**Droping some Column**

```python
df.drop(columns = ['class', 'who', 'adult_male', 'deck', 'embark_town',
       'alive', 'alone'],inplace = True)
```

```python
df.head(5)
```

**Now i am Spliting the data**

```python
X_train ,X_test,y_train,y_test =  train_test_split(df.drop(columns=['survived']),df['survived'],test_size = 0.2, random_state= 42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

**1st part of pipeline : SimpleImpute**

```python
#impute Transformer
# For age column we are using MEAN
# If we do not use the passthrough then our other column will be drop
# [2] -> use column index insted on their name becasue if we use name of the column then it might be possible the code will be break down in the pipeline .
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

**2st part of pipeline : OnehotEncoding**

```python
#One Hot Encoding
#We can give multiple column within the same object i.e, i will give sex and embaked within the same object

trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

**3st part of pipeline : Scaling**

```python
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

**4st part of pipeline : Feature Selection**

```python
trf4 = SelectKBest(score_func=chi2,k=8)
```

**5st part of pipeline : Model Selection**

```python
trf5 = DecisionTreeClassifier()
```

**Main Piepline**
1.   Create pipeline
2.   make_pipeline




**Pipeline**:

You manually specify a name for each step in the pipeline.
Useful when you need more control and want to reference steps later by their names.

**(Always use) make_pipeline**:
Automatically generates names for steps based on the classes of the transformers or models used.
Easier and faster for prototyping when you don’t need custom step names.

**Key Difference**:
Pipeline gives explicit control over step names.
make_pipeline simplifies the process by auto-generating names.








```python
#Create pipeline
pipe = Pipeline([
    ('P1',trf1),
    ('P2',trf2),
    ('P3',trf3),
    ('P4',trf4),
    ('P5',trf5)
])
```

```python
pipe_make = make_pipeline(
    trf1,
    trf2,
    trf3,
    trf4,
    trf5
)
```

```python
pipe.fit(X_train,y_train)
```

**Explore the pipeline**

```python
pipe.named_steps
```

```python
from sklearn import  set_config
set_config(display='diagram')
```

```python
#Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

**Cross validation Using Pipelines**

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```
