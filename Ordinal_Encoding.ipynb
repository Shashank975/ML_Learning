```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head(5)
```

Now we will do the train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size = 0.3 )
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']])
```

We Do fitting (fit) on the train (X_train) column only and transfrorm both (X_train,y_train)  

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
